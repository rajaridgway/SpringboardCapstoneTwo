## <div class="alert alert-block alert-success">Summary of Insights</div>

**How different are schoolsin terms of academic performance?**
 - Schools appear to have a range of scores for both SAT and ACT performance. **For SAT, schools range from an average of 795 to 1329. For ACT, schools range from 13.9 to 29.8.**
 - For the ACT exam, the scores skew left. For the SAT, the scores do not appear to have a skew.
 
**How different are schoolsin terms of attendance?**
 - Attendance is captured in three ways with the percent of students missing 0-5 days, 6-15 days, and 16+ days. GOSA assigns schools indicator ratings based on the percentage of students with 16+ absent days: exemplary (5% or fewer), acceptable (>5% and <15%), and unacceptable (>15%).
 - **Of the 344 schools in the dataset, 36 have exemplary attendance, 216 have acceptable attendance, and 92 have unacceptable attendance.**
 - Another attendance metric is chronic absenteeism, which is when a student misses more than 10% of the school days. Asian and Native American students are are likely to have lower rates of chronic absenteeism than students of other races. Similarly, students without disabilities are more likely to have lower rates of chronic absenteeism than students with disabilties. The same is true for students who are not economically disadvantaged.  
 
**How different are schools in terms of enrollment?**
 - The two most common races represented are White (school average of 41%) and Black (school average of 39%). **However, there are some schools with 0% Black students and some with 0% White students. Additionally, some schools have 97% Black students while others have 94% White students.** Native American students are the least represented with the average school have less than 1% Native American students. On average, schools have approximately 3% Asian students and 13% Hispanic students.
 - Most schools have very few students designated as migrants and most schools have fewer than 10% of students designated as Limited English Proficiency (a few have greater than 50%). 
 - Most schools have around 15% students with disabilities and none with more than 25%.
 - **There is a large variation in how many students with an economically disadvantaged desgination are at a school.** 

**Are there any schools that out- or under-perform the average with ACT or SAT?**
 - For both the ACT and SAT, there appears to be a single school that is higher performing. After filtering for the single school, it is **[Gwinnet School of Mathematics, Science, and Technology](https://www.gcpsk12.org/GSMST)**. According to their website, it has been the top ranked high school in GA since 2013. Students are chosen to attend by random lottery if they meet eligibility criteria (unknown); it was a public charter school, but in 2016-2017 became a traditional public school in order to impose an [admission requirement](https://www.ajc.com/blog/get-schooled/longer-charter-gwinnett-school-mathematics-science-and-technology-will-require-algebra-for-admission/IMIZISnjgsn2FHhLWgLQxH/).

**Does performance on SAT equate to performance on ACT?**
 - There is a high level of positive correlation between the SAT and ACT scores at the school level (R^2 = 0.83). There is also a positive correlation between the SAT and ACT at the district level (R^2 = 0.68), but not to the same extent.
 
**How do schools perform in comparison to college-ready benchmarks?**
 - Both the ACT and the SAT have benchmark scores that demonstrate that a student is "college-ready." For the ACT, the benchmark score for the overall score (21.25) is calculated by taking the average of the benchmarks for English (18), Mathematics (22), Reading (22), and Science (23). For the SAT, the benchmark for the composite score is simply 1010. 
 - For this dataset of highschools in GA, 94 schools were at or above the ACT benchmark and 180 were at or above the SAT threshold. **This large difference is an interesting result given the high level of positive correlation between the SAT and ACT scores at the school level.** 
 - Also note that the average national ACT score was 20.6 and the average GA ACT score was 20.8, which are both below the ACT benchmark for college readiness. The average GA SAT score was 1028, which is above the SAT benchmark for college readiness.

```python
# Define a function for a more customized visual (adapted from: https://mode.com/example-gallery/python_histogram/)
def simple_hist(df, column, xlabel, ylabel, benchmark):
    
    ax = df.hist(column=column, bins=25, grid=False, figsize=(12,8), color='#86bf91', zorder=2, rwidth=0.9)

    ax = ax[0]
    for x in ax:

        # Despine
        x.spines['right'].set_visible(False)
        x.spines['top'].set_visible(False)
        x.spines['left'].set_visible(False)

        # Switch off ticks
        x.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")

        # Draw horizontal axis lines
        vals = x.get_yticks()
        for tick in vals:
            x.axhline(y=tick, linestyle='dashed', alpha=0.4, color='#eeeeee', zorder=1)

        # Remove title
        x.set_title("")

        # Set x-axis label
        x.set_xlabel(xlabel, labelpad=20, weight='bold', size=12)

        # Set y-axis label
        x.set_ylabel(ylabel, labelpad=20, weight='bold', size=12)

        # Format y-axis label
        x.yaxis.set_major_formatter(StrMethodFormatter('{x:,g}'))

        
    # Add benchmark
    plt.axvline(x=benchmark)
    
    return ax
```

```python
simple_hist(scores, 'INSTN_AVG_SCORE_VAL_act', 'ACT Scores', 'Schools', 21.25)
```

### ACT vs. SAT at School-Level

```python
# Basic plot of ACT vs. SAT
scores.plot.scatter(x='INSTN_AVG_SCORE_VAL_act', y='INSTN_AVG_SCORE_VAL_sat')
plt.show()
```

```python
# Add trend line (adapted from https://stackoverflow.com/questions/41635448/how-can-i-draw-scatter-trend-line-on-matplot-python-pandas)
x = scores['INSTN_AVG_SCORE_VAL_act']
y = scores['INSTN_AVG_SCORE_VAL_sat']

plt.scatter(x,y)
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
y_hat = np.poly1d(z)(x)
plt.plot(x,p(x),"r--")


text = f"$y={z[0]:0.3f}\;x{z[1]:+0.3f}$\n$R^2 = {r2_score(y,y_hat):0.3f}$"
plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     fontsize=14, verticalalignment='top')

plt.show()
```
